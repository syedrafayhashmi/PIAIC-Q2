```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
iris = iris.sample(frac = 1)
iris.tail()
```

```python
iris.shape
```

```python
iris_new = np.array(iris.drop(["variety"], axis = 1))
iris_new
```

```python
iris_new.shape
```

```python
variety = np.array(iris["variety"])
variety
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
variety = le.fit_transform(variety)
variety
```

```python
from tensorflow.keras.utils import to_categorical
enc_variety = tf.keras.utils.to_categorical(variety)
enc_variety
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris_new, enc_variety, test_size = 0.2, random_state = 42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(12, input_shape = (4,)))
model.add(Dense(8, activation = "relu"))
model.add(Dense(3, activation = "softmax"))
```

```python
model.compile(optimizer = "adam",
             loss = "categorical_crossentropy",
             metrics = ["acc"])
```

```python
model.fit(x_train, y_train, epochs = 50, validation_split = 0.1, batch_size = 12, verbose = 2)
```

```python
evaluation = model.evaluate(x_test, y_test, verbose = 2)
print("Loss: ",evaluation[0])
print("Accuracy: ",evaluation[1])
```

```python
i = 8
prediction =  model.predict_classes(x_test[i].reshape(1,4))
```
