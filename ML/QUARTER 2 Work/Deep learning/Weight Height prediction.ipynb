```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("datasets/weight-height.csv")
df.head()
```

```python
print("Missing Total values: ",df.isna().sum().sum())
```

```python
plt.hist(x=df.Height, color="skyblue")
plt.title("Univariate: Height Histogram")
plt.xlabel("Height")
plt.ylabel("Total Counts")
plt.plot();
```

```python
plt.hist(x=df.Weight, color="c", bins = 10)
plt.title("Univariate: Weight Histogram")
plt.xlabel("Weight")
plt.ylabel("Total Counts")
plt.plot();
```

```python
# KDE distribution for Height
df.Height.plot(kind="kde", title='Univariate: Height KDE', color='c');
```

```python
# KDE distribution for Weight
df.Weight.plot(kind="kde", title='Univariate: Weight KDE', color='c');
```

```python
plt.scatter(x=df["Height"], y=df["Weight"], color='c')
plt.title("Bivariate: Height Vs Weight Using Matplotplib")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.plot();
```

```python
df.head()
```

```python
height = df.iloc[:, 1:2].values
weight = df.iloc[:, 2:3].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(height, weight, test_size = 0.3, random_state = 31)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()

model_fit = regressor.fit(x_train, y_train)
```

```python
y_predict = regressor.predict(x_test)
```

```python
y_predict
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_train, regressor.predict(x_train), color='r')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Heigth Vs Weight Prediction")
plt.show()
```

```python
from sklearn.metrics import r2_score

print(f"Model Accuracy is: {regressor.score(x_test, y_test)}")
```

```python
r2_score(y_test, y_predict)
```
